```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1.0, 4.5, endpoint=True)
y = np.linspace(-1.5, 4.5, endpoint=True)

X,Y = np.meshgrid(x,y)
```

```python
D1 = np.sqrt(X*X + Y*Y)
D2 = np.sqrt((1-X)*(1-X) + (1-Y)*(1-Y))
D3 = np.sqrt(X*X + (4-Y) * (4-Y))
```

```python
Z1 = np.sqrt(D1) + np.sqrt(D2)
Z2 = np.sqrt(D1) + np.sqrt(D3)
Z3 = np.sqrt(D2) + np.sqrt(D3)

allz = np.empty((3, Z1.shape[0], Z1.shape[1]))
allz[0] = Z1
allz[1] = Z2
allz[2] = Z3
Z = np.min(allz, axis=0)
```

```python
rows, cols = Z.shape
```

```python
fig, ax = plt.subplots()
contour = ax.contourf(X,Y,Z, 50)
fig.colorbar(contour)
plt.show()
```

```python
Z1 = D1 + D2
Z2 = D1 + D3
Z3 = D2 + D3

allz = np.empty((3, Z1.shape[0], Z1.shape[1]))
allz[0] = Z1
allz[1] = Z2
allz[2] = Z3
Z = np.min(allz, axis=0)
```

```python
fig, ax = plt.subplots()
contour = ax.contourf(X,Y,Z, 50)
fig.colorbar(contour)
plt.show()
```
